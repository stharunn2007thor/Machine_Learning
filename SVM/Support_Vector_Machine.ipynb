```python

from sklearn.datasets import load_iris
iris = load_iris()
df  = pd.DataFrame(iris.data , columns= iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df[df['target'] ==2].head()
```

```python
df0  = df[df['target'] ==0]
df1  = df[df['target'] ==1]
df2  = df[df['target'] ==2]
```

```python
import matplotlib.pyplot as plt

plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker ='+',color = 'green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker ='+',color = 'blue')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],marker ='+',color = 'red')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('Length vs Width')
plt.legend(['setosa','versicolor','virginica'])
plt.show()
```

```python
df['flower_names'] = df['target'].apply(lambda x: iris.target_names[x])
df.head()
```

```python
X = df.drop(['flower_names', 'target'], axis=1)
y = df['target']
print(X.head())
```

```python
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn import model_selection
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
model.score(X_train,y_train)

```
